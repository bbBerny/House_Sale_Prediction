```python
# Import main libraries and initial configurations

import numpy as np #type: ignore
import pandas as pd # type: ignore
import matplotlib.pyplot as plt #type: ignore
import seaborn as sns #type: ignore

%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
# Open and read the training set
dataset = pd.read_csv('Data/train.csv')

# Print the shape of the dataset
print(dataset.shape)
```

```python
# Dataset top 5 records
dataset.head()
```

### 1. Missing values

```python
# Identify features with missing values

# If a feature (column) has at least one missing (NaN) value, include it in the list
# This will create a list of column names with missing values
features_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0]

# For each feature with missing values, print the name and the proportion of missing data
# Since isnull() returns True (1) for NaNs and False (0) for non-NaNs,
# the mean() gives the percentage of missing values in the column
# The result is rounded to 4 decimal places
for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

#### Since there are many missing values, we need to find the relationship between missing values and sales price

- I want to find a relationship between my dependent feature (Sales price) and all my other features

```python
for feature in features_with_nan:
    data = dataset.copy()

    # Create a binary indicator for missing values:
    # 1 if the value is missing (NaN), 0 otherwise
    # This is helpful to visualize how missingness affects SalePrice
    data[feature] = np.where(data[feature].isnull(), 1, 0)

    # Plot the median SalePrice for rows with and without missing values
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.ylabel('Median SalePrice')
    plt.xlabel('Missing Indicator (1 = Missing)')
    plt.show()

    # Interpretation: If there is a clear difference between the bars,
    # then the presence of missing values in this feature may be informative (not missing at random)
```

### 2. Numerical variables

```python
# Obtain all numerical features, essentially all distinct from a object datatype
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print(f'Number of numerical features: {len(numerical_features)}')
```

```python
dataset[numerical_features].head()
```

Temporal variables (Ex. Datetime variables):
- From the dataset we have 4 year variables. We have to extract information from the datetime like number of years, or number of days.
- One example in this scenario can be a difference in years between the year the house was built and the year the house was sold.
- We will be performing this analysis in the feature engineering section

```python
# List of variables that contain year information

year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# Let us explore the content of these year variables
# These are the unique years that appear in the dataset
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Is there a relationship between YrSold and the SalePrice? 

# Group the dataset by the year the house was sold (YrSold),
# compute the median SalePrice for each year,
# and plot the result to observe potential trends over time.
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Sale price')
plt.title('House price vs Year Sold')

# Interpretation: The prize is decreasing over time, is this actually accurate?
```

```python
year_feature
```

```python
# Now we will compare the difference between All years feature with salePrice

for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        # We will try and capture the difference between year variable and
        # the year the house was sold
        data[feature] = data['YrSold'] - data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

- Numerical variables are usually of 2 types:
    - Continuous variables, they have a infinite set of values
    - Discret variables, fixed set of integer values

```python
threshold = 25
# If a feature in numerical features has more than 25 unique values, then we will consider it as a continuous variable, else we will trat it as a discrete variable 
# Also we dont want to analyze features that contain year information and also ID
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique()) < threshold and feature not in year_feature + ['Id']]
print(f'Discrete variables count: {len(discrete_feature)}')
```

```python
# Let us see the head for the discrete features
dataset[discrete_feature].head()
```

```python
# Lets try to find the relationship between the discrete features and sale price
# In EDA we will always want to gather more information about the data so it is crucial for us to compare the Sale Price with the independent features

for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()

# Interpreatation: I can see a monotonic relationship between the overall quality of the house and the Sale Price
# Monotonic is a type of relationhip between two variables where the direction of the relationship is consistent, but not necessarily linear
```

Continuous variables

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_feature and feature not in year_feature and feature not in ['Id'] ]
print(f'Number of continuous features: {len(continuous_features)}')
```

```python
# Since we are working with continuous values, let us try to find a distribution, normally done with histograms

for feature in continuous_features:
    data = dataset.copy()
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()

# Some of the features have a gaussian distribution, but some dont. This is basically prove that we are working with skewed data
```
